```python
# Import pandas 
import pandas as pd 
  
# Read csv file  
vine_table_to_load = "vine_table.csv"
vine_table_df = pd.read_csv(vine_table_to_load)

# Print it out 
vine_table_df.head()
```

```python
#filtering for total votes that are 20 and up
vine_table_df = vine_table_df[(vine_table_df.total_votes > 19)]
vine_table_df.head()
```

```python
#confirming above code worked as intended
vine_table_df.sort_values(by=["total_votes"]) 
```

```python
#filtering for data that have more than 50% helpful votes
vine_table_df = vine_table_df[(vine_table_df.helpful_votes / vine_table_df.total_votes > 0.5 )]
vine_table_df.head(10)
```

```python
#confirming the count of all reviews
len(vine_table_df.total_votes)
```

```python
#new dataframe for vine participates
vine_table_Y_df = vine_table_df[(vine_table_df.vine == "Y")]
vine_table_Y_df.head(5)
```

```python
#total number of reviews with vine program
len(vine_table_Y_df.total_votes)
```

```python
#total number of 5 star reviews with vine program
five_star_Y_df = vine_table_Y_df[(vine_table_Y_df.star_rating == 5)]
len(five_star_Y_df)
```

```python
#new dataframe for non-vine participates
vine_table_N_df = vine_table_df[(vine_table_df.vine == "N")]
vine_table_N_df.head(5)
```

```python
#total number of reviews without vine program
len(vine_table_N_df.total_votes)
```

```python
#total number of 5 star reviews without vine program
five_star_N_df = vine_table_N_df[(vine_table_N_df.star_rating == 5)]
len(five_star_N_df)
```

```python
#percentage of 5 star reviews based on vine and not vine
#vine 
vine_percentage = round(len(five_star_Y_df) / len(vine_table_Y_df.total_votes),2)
print(vine_percentage)
```

```python
#not vine
not_vine_percentage = round(len(five_star_N_df)/len(vine_table_N_df.total_votes),2)
not_vine_percentage
```
